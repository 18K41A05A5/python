```
month=input("enter the month:")
winter=["December","January","February"]
spring=["March","April","May"]
summer=["June","July","August"]
autumn=["September","October","November"]
if month in winter:
  print("winter")
elif month in spring:
  print("spring")
elif month in summer:
  print("summer")
elif month in autumn:
  print("autumn")
```
